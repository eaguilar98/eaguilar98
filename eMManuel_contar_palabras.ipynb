```python
import csv
import pandas as pd
import numpy as np
import urllib.request
import json
import requests    
import pprint
```

```python
dic1={
    ord(","):None,
    ord("."):None,
    ord("-"):None,
    ord("@"):None,
    ord(":"):None,
    ord(";"):None,
    ord("_"):None,
    ord("+"):None,
    ord("#"):None,
    ord("["):None,
    ord("]"):None,
    ord("*"):None,
    ord("/"):None,
    ord("{"):None,
    ord("?"):None,
    ord(""):None,
    ord("("):None,
    ord(")"):None,
    ord("!"):None,
    ord("\\"):None
    
    
}
```

```python
mi_libro="https://www.gutenberg.org/cache/epub/67596/pg67596.txt"

webLibro=urllib.request.urlopen(mi_libro).read().decode('utf-8').lower()
webLibro=webLibro.translate(dic1)
#print(webLibro[:5000])
```

```python
palabras=webLibro[:].split()
```

```python
cuenta_palabras={}
```

```python
estas_no=[
          "that",
          "with",
          "said",
          "they",
          "from",
          "were"
          "then",
          "when",
          "this",
          "there",
          "would",
          "have",
          "their",
          "been",
          "over",
          "could",
          "into",
          "down",
          "them",
          "were",
          "then",
          "which",
          "what",
          "just",
          "some",
          "last",
          "your"
]
```

```python
suma=0
for x in palabras:
    if len(x)>3 : 
         if x not in estas_no:       
            for y in range(len(palabras)):
                if(x==palabras[y]):
                    suma=suma+1
            cuenta_palabras[x]=suma
            suma=0
```

```python
palabras_df=pd.DataFrame(list(cuenta_palabras.items()), columns=["palabra","cantidad"])
```

```python
palabras_df=palabras_df.sort_values("cantidad", ascending=False)
```
